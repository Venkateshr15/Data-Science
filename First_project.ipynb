```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from google.colab import files
```

```python
df=pd.read_csv('/content/movie_info.csv')
```

```python
pd.read_csv('/content/movie_info.csv')
```

```python
user_preferences = {
    'unknown': 0,
    'Action': 1,
    'Adventure': 0,
    'Animation': 0,
    "Children's": 0,
    'Comedy': 0,
    'Crime': 0,
    'Documentary': 0,
    'Drama': 0,
    'Fantasy': 0,
    'Film-Noir': 0,
    'Horror': 0,
    'Musical': 0,
    'Mystery': 0,
    'Romance': 0,
    'Sci-Fi': 1,
    'Thriller': 1,
    'War': 0,
    'Western': 0
}
```

```python
genre_columns = list(user_preferences.keys())
user_vector = np.array([user_preferences[genre] for genre in genre_columns])
```

```python
movie_genres = df[genre_columns].values
```

```python
similarities = cosine_similarity([user_vector], movie_genres)[0]
```

```python
df['similarity'] = similarities
top_recommendations = df.sort_values(by='similarity', ascending=False)[['movie title', 'similarity']].head(10)
```

```python
print("Top 10 Personalized Movie Recommendations:")
print(top_recommendations.to_string(index=False))
```
